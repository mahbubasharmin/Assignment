```python
# Upload the Gym data in google drive
gym = pd.read_csv('/content/drive/MyDrive/Darasets/gym_membership.csv')
```

```python
# Display last 5 rows
print("\nLast 5 rows of the dataset:")
print(gym.tail())
```

```python
# Check data types
print("\nData Types:")
print(gym.dtypes)
```

```python
#task-1
# 1. Bar chart of gym membership types
gender_count = gym['gender'].value_counts()
plt.bar(['Female', 'Male'], gender_count.values,color=['green', 'blue'])
plt.title('Gender Count')
plt.show()
```

```python
# 2. Histogram of time in gym
plt.hist(gym['avg_time_in_gym'])
plt.title('Average Time in Gym')
plt.show()
```

```python
# 3. Scatter plot: Age vs visits per week
plt.scatter(gym['Age'],gym['visit_per_week'])
plt.title('Age and visit per week')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Gender vs Age')
plt.show()
```

```python
# 2. Count plot: Days per week
sns.countplot(data=gym,x='days_per_week')
plt.title('Days per Week')
plt.show()
```

```python
# 3. Bar plot: Average visits by membership type
sns.barplot(data=gym,x='abonoment_type',y='visit_per_week')
plt.title('Average visits by membership')
plt.show()
```
